```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### 1.Binary Thresholding

```python
# Load the image
image = cv2.imread('../images/lenna.png', 0)

#Define a threshold value
binary_image = image > 127
binary_image
```

```python
# Show the original and binary images
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')
plt.subplot(1, 2, 2)
plt.title('Binary Image')
plt.imshow(binary_image, cmap = 'gray');
```

```python
print(binary_image.dtype)

binary_image = binary_image.astype(np.uint8) * 255
print(binary_image)

cv2.imshow('Binary Images', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2.Inverse Binary Thresholding

```python
# Load the image
image = cv2.imread('../images/lenna.png', 0)

#Define a threshold value
binary_image = image < 127
binary_image
```

```python
# Show the original and inverse binary images
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')
plt.subplot(1, 2, 2)
plt.title('Inverse Binary Image')
plt.imshow(binary_image, cmap = 'gray');
```

```python
print(binary_image.dtype)

binary_image = binary_image.astype(np.uint8) * 255
print(binary_image)

cv2.imshow('Binary Images', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, thresh1 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize = (16, 8))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')
plt.subplot(1, 3, 2)
plt.title('Binary Image')
plt.imshow(thresh1, cmap = 'gray')
plt.subplot(1, 3, 3)
plt.title('Inverse Binary Image')
plt.imshow(thresh2, cmap = 'gray');
```
