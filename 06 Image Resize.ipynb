```python
import cv2
from PIL import Image
from skimage.transform import resize
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread('../images/lenna.png')

# Resize the image to a fixed width and height
resized_image = cv2.resize(image, (100, 200))

# show the rotated image
cv2.imshow("Resized image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Resize using scaling factor

```python
# Load the image
image = cv2.imread('../images/lenna.png')

# Resize the image with aspect ratio preserved
resized_image = cv2.resize(image, None, fx=0.5, fy=0.75)

# show the rotated image
cv2.imshow("Resized image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load the image
image = cv2.imread('../images/lenna.png')

# Resize the image with aspect ratio preserved
resized_image = cv2.resize(image, None, fx=1.5, fy=1.75)

# show the rotated image
cv2.imshow("Resized image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Resize using PIL

```python
# Load the image
image = Image.open('../images/lenna.png')

# Resize the image to a fixed size
resized_image = image.resize((800, 800))

# Resize the image with aspect ratio maintained
resized_image = image.resize((800, 600))

# show the resized image
resized_image
```

### Resizing using scikit-image

```python
# Load the image
image = plt.imread('../images/lenna.png')

# Resize with bilinear interpolation
resized_image = resize(image, (1000, 1200), anti_aliasing=True)

# Resize with bicubic interpolation
resized_image = resize(image, (1300, 1300), anti_aliasing=True, order=3)

# show the rotated image
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resized_image)
```
